```python
#first plot
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,1,10)
m=0.6
c=25
y=m*x+c
plt.plot(x,y)
plt.show()
```

```python
#subplot
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y=6+3*np.sin(2*x)
fig, a=plt.subplots()
a.plot(x,y,linewidth="0.5",color="purple")
a.set(xlim=(0,9), xticks=np.arange(1,9),
      ylim=(0,9), yticks=np.arange(1,9))
plt.show()
```

```python
#scatterplot
import matplotlib.pyplot as plt
import numpy as np
temp=np.array([10,20,25,30,40,50,60,70,80,90,100,110,120])
res=np.array([18476,12185,10000,8260,5740,4080,2960,2188,1645,1257,973,765,608])
plt.scatter(temp,res,marker="x")
plt.show()
```

```python
#garnishing
import matplotlib.pyplot as plt
import numpy as np
temp=np.array([10,20,25,30,40,50,60,70,80,90,100,110,120])
res=np.array([18476,12185,10000,8260,5740,4080,2960,2188,1645,1257,973,765,608])
plt.plot(temp,res,marker="x",linestyle="dotted")
plt.title("thermistor characteristics")
plt.xlabel("temperature -->")
plt.ylabel("resistance -->")
plt.grid()
plt.show()
```

```python
#bar and pie graph
import matplotlib.pyplot as plt
import numpy as np
arr1=np.array(["a","b","c","d"])
arr2=np.array([10,12,25,15])
plt.bar(arr1,arr2)
plt.show()
plt.pie(arr2)
plt.show()
```
