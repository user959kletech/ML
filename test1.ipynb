```python
x=input("enter the string ")
y=len(x)
z=y-1
p=0
for i in range(0,y,1):
  if(x[i] is not x[z-i]):
    p=1
    break
if(p==1):
  print("the string is not palindrome")
else:
  print("the string is palindrome")
```
