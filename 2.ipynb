```python
#bitwise shift op
a=5
a=a<<1
print(a)
a=a>>2
print(a)
```

```python
#while loop
list1=[{'ramesh',1,'hubli'},{'suresh',2,'dharwad'},{'ram',3,'gadag'},{'shyam',4,'bengaluru'}]
x=0
while x<4:
  print(list1[x])
  x=x+1
```

```python
#for loop
list1=[{'ramesh',1,'hubli'},{'suresh',2,'dharwad'},{'ram',3,'gadag'},{'shyam',4,'bengaluru'}]
for x in range(0,4,1):
  print(list1[x])
```
